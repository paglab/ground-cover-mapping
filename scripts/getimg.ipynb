```python
import os
import requests

def google_search_images(query, api_key, cse_id, num_images=10):
    search_url = "https://www.googleapis.com/customsearch/v1"
    params = {
        "q": query,
        "cx": cse_id,
        "key": api_key,
        "searchType": "image",
        "num": 10,  # 每次查询返回的图片数量（最大值为10）
        "start": 8  # 查询的起始位置
    }
    image_urls = []
    
    while len(image_urls) < num_images:
        response = requests.get(search_url, params=params)
        results = response.json()
        if "items" in results:
            for item in results["items"]:
                image_urls.append(item["link"])
                if len(image_urls) >= num_images:
                    break
        else:
            break
        
        params["start"] += 10
    
    return image_urls[:num_images]

def download_images(image_urls, save_dir):
    os.makedirs(save_dir, exist_ok=True)
    for i, url in enumerate(image_urls):
        try:
            img_data = requests.get(url).content
            img_name = os.path.join(save_dir, f"image_{i+1}.jpg")
            with open(img_name, "wb") as img_file:
                img_file.write(img_data)
            print(f"Downloaded {img_name}")
        except Exception as e:
            print(f"Failed to download {url}: {e}")

# 使用你的API密钥和搜索引擎ID
API_KEY = "AIzaSyDjmeV1qvW3CK00TWEKC8H9L55Fmh--z0g"
CSE_ID = "616740f206f65431e"

keywords = ['Grass', 'Herb', 'Litter', 'Soil', 'Stone', 'Wood', 'Woodchip']

for keyword in keywords:
    image_urls = google_search_images(keyword, API_KEY, CSE_ID, num_images=100)
    download_images(image_urls, save_dir=keyword)
```
