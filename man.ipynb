```python
#cut into small tiles because of memory limitation
from PIL import Image
import os

def crop_image(image_path, output_dir, crop_size=(640, 640)):
    """
    将图像裁剪为多个指定大小的小块并保存。
    
    :param image_path: 原始大图像的路径
    :param output_dir: 裁剪后的小图像的保存目录
    :param crop_size: 裁剪块的大小，默认是(640, 640)
    """
    # 创建保存小图像的目录
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    # 打开原始图像
    image = Image.open(image_path)
    image_width, image_height = image.size
    
    crop_width, crop_height = crop_size

    # 计算可以在宽度和高度方向裁剪的块数
    num_cols = image_width // crop_width
    num_rows = image_height // crop_height

    # 逐块裁剪并保存
    for row in range(num_rows):
        for col in range(num_cols):
            left = col * crop_width
            upper = row * crop_height
            right = left + crop_width
            lower = upper + crop_height

            # 裁剪图像
            cropped_image = image.crop((left, upper, right, lower))

            # 保存裁剪后的图像
            cropped_image_filename = f"crop_{row}_{col}.png"
            cropped_image_path = os.path.join(output_dir, cropped_image_filename)
            cropped_image.save(cropped_image_path)
            print(f"保存 {cropped_image_path}")

    print("裁剪完成。")

# 使用示例
image_path = "raw\\img.png"  # 输入大图像的路径
output_dir = "cropped"  # 小图像的保存路径
crop_image(image_path, output_dir)
```

```python
#Restore imgs
from PIL import Image
import os

def stitch_images(input_dir, output_image_path, original_size, crop_size=(640, 640)):
    """
    将裁剪的小图像拼接回原始大小的图像。

    :param input_dir: 小图像所在的目录
    :param output_image_path: 拼接后完整图像的保存路径
    :param original_size: 原始大图像的尺寸（宽度，高度）
    :param crop_size: 小图像的尺寸，默认是(640, 640)
    """
    # 原始图像的尺寸
    original_width, original_height = original_size
    
    crop_width, crop_height = crop_size

    # 计算原图可以分成多少块
    num_cols = original_width // crop_width
    num_rows = original_height // crop_height

    # 创建一个新的空白图像，用于存放拼接结果
    stitched_image = Image.new('RGB', (original_width, original_height))

    # 按行列顺序读取小图像并拼接
    for row in range(num_rows):
        for col in range(num_cols):
            # 构建小图像文件名
            cropped_image_filename = f"crop_{row}_{col}.png"
            cropped_image_path = os.path.join(input_dir, cropped_image_filename)
            
            # 打开小图像
            cropped_image = Image.open(cropped_image_path)
            
            # 计算放置该小图像的位置
            left = col * crop_width
            upper = row * crop_height
            stitched_image.paste(cropped_image, (left, upper))
            print(f"拼接 {cropped_image_filename} 到位置 ({left}, {upper})")

    # 保存拼接后的图像
    stitched_image.save(output_image_path)
    print(f"拼接完成，保存为 {output_image_path}")

# 使用示例
input_dir = "cropped_out"  # 已预测的小图像目录
output_image_path = "stitched_image.png"  # 输出完整图像的路径
original_size = (7683, 10689)  # 原始大图像的尺寸，假设为10240x7680像素
stitch_images(input_dir, output_image_path, original_size)
```
